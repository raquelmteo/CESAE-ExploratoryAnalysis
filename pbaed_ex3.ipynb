```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import warnings 
warnings.filterwarnings("ignore")
```

Read the CSV file

```python
data = pd.read_csv('C:/Users/raque/OneDrive/Documentos/Python/Análise Exploratória de Dados/Trabalho Final/pbaed_ex3.csv')
```

Show data from the CSV file

```python
print(data)
```

Represents the data in a Scatter Plot

```python
plt.scatter(data['Distancia'], data['Preco'])
plt.xlabel('Distance')
plt.ylabel('Price')
plt.title('Scatter Plot')
plt.show()
```

Detailed Correlation

```python
X = data['Distancia']
y = data['Preco']

meanX = np.mean(X)
meany = np.mean(y)

difX = X - meanX
dify = y - meany

num = np.sum(difX * dify)
den = np.sqrt(np.sum(difX ** 2) * np.sum(dify ** 2))

correlation = num/den

print(f"""Correlation: {round(correlation, 2)}
A positive value indicates that there is a positive relation between the two variables. When the distance increases, the price also tends to increase.""")

plt.scatter(data['Distancia'], data['Preco'])
plt.title('Relation Distance and Price')
plt.xlabel('Distance')
plt.ylabel('Price')
plt.show()
```

Correlation

```python
print(f"""Correlation: {round(data['Distancia'].corr(data['Preco']),2)}
A positive value indicates that there is a positive relation between the two variables. When the distance increases, the price also tends to increase.""")

plt.scatter(data['Distancia'], data['Preco'])
plt.title('Relation Distance and Price')
plt.xlabel('Distance')
plt.ylabel('Price')
plt.show()
```

Estimated price for the distance of 1200km with detailed Linear Regression

```python
X = data['Distancia']
y = data['Preco']

meanX = np.mean(X)
meany = np.mean(y)

difX = X - meanX
dify = y - meany

prod_dif = difX * dify

sum_prod = np.sum(prod_dif)
sum_dif = np.sum(difX**2)

b = sum_prod / sum_dif
a = meany - b * meanX

print(f"Slope Coefficient: {round(a, 2)}")
print(f"Intercept: {round(b, 2)}")

print("Linear Regression Equation: y =", round(b, 2), "x +", round(a, 2))

prec = b * 1200 + a
print(f"Estimated price for 1200 km: {round(prec)}")

plt.scatter(X, y)
plt.plot(X, b * X + a)
plt.title('Linear Regression - Distance vs. Price')
plt.xlabel('Distance')
plt.ylabel('Price')
plt.show()


```

Estimated price for the distance of 1200km with Linear Regression

```python
valuesX = data[['Distancia']]
valuesy = data['Preco']

model = LinearRegression()
model.fit(valuesX, valuesy)

prec = model.predict([[1200]])[0]
print(f"Estimated price for 1200 km: {round(prec)}")

plt.scatter(X, y)
plt.plot(X, b * X + a)
plt.title('Linear Regression - Distance vs. Price')
plt.xlabel('Distance')
plt.ylabel('Price')
plt.show()
```

Detailed R-square

```python
X = data['Distancia']
y = data['Preco']

meanX = np.mean(X)
meany = np.mean(y)

difX = X - meanX
dify = y - meany

num = np.sum(difX * dify)
den = np.sqrt(np.sum(difX**2) * np.sum(dify**2))

r = num/den

r2 = r ** 2

print(f"""Coefficient of determination: {round(r2, 2)}
A positive value indicates that there is a positive relation between the two variables. When the distance increases, the price also tends to increase.""")
```

R-square

```python
X = data[['Distancia']]
y = data['Preco']

model = LinearRegression()
model.fit(X, y)

slope = model.coef_[0]
intercept = model.intercept_

r2 = model.score(X,y)
print(f"Coefficient of determination: {round(r2, 2)}")

regression_line = slope * X + intercept

plt.scatter(X, y)
plt.plot(X, regression_line)
plt.xlabel('Distance')
plt.ylabel('Price')
plt.show()
```
