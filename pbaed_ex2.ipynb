Standard Deviation Sellers

```python
print(f"""The standard deviation of the respective sellers are: 
      Salesperson 1: {(round(data['Vendedor1'].std(),2))} 
      Salesperson 2: {(round(data['Vendedor2'].std(),2))}
      Salesperson 3: {(round(data['Vendedor3'].std(),2))}
      Salesperson 4: {(round(data['Vendedor4'].std(),2))} 
With salesperson 1 being the lowest standard deviation, so the most consistent in the last six months.""")
```

Create a boxplot for each of the salespeople

```python
data_vendedores = data.drop(columns=['Mes'])

fig, axes = plt.subplots(1, 4, figsize=(15, 5))

for i, Salesperson in enumerate(data_vendedores.columns):
    data_vendedores.boxplot(column=[Salesperson], ax=axes[i])
    axes[i].set_title(Salesperson)

plt.tight_layout()
plt.show()
```
